```python
import pandas as pd
import numpy as np
```

```python
df_glcm = pd.read_csv('glcm_dataset.csv')
df_glcm = df_glcm.drop(columns=['Unnamed: 0', 'label'])
df_glcm = df_glcm.rename(columns={
    'img_ids': 'glcm_id'
})

df_color = pd.read_csv('color_dataset.csv')
df_color = df_color.drop(columns=['Unnamed: 0'])
df_color = df_color.rename(columns={
    'img_ids': 'color_id'
})
```

```python
merged = pd.merge(df_glcm, df_color, left_on='glcm_id', right_on='color_id',
                  how='inner').drop('color_id', axis=1).rename(columns={'glcm_id': 'id'})
col = merged.pop("id")
merged.insert(0, col.name, col)
merged
```

```python
merged.to_csv('merged_features.csv')
```
